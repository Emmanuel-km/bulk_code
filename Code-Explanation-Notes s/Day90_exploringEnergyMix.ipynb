```python
import seaborn as sns

import matplotlib.pyplot as plt

# Create dummy data
data = {'Category': ['Biofuels and Waste', 'Coal', 'Oil', 'Solar,wind and other renewable','Hydropower'], 'Values': [62.2,3.6, 16.7, 16.4,1]}

# Create pie chart
plt.figure(figsize=(8, 6),facecolor="#0f91258d")
colors = sns.color_palette("husl", len(data['Category']))
plt.pie(data['Values'], labels=data['Category'], autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Energy mix ', fontsize=15, color='white')
plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

Values = [0.27, 0.0, 0.0, 3.6200000000000006, 0.85, 5.55, 0.45999999999999996, 1.8099999999999996]
Category = ['Bioenergy', 'Coal', 'Gas', 'Hydro', 'Other fossil', 'Other renewables', 'Solar', 'Wind']

# Remove zero values and corresponding categories (optional - you can keep this or comment out)
# This removes categories with zero values for cleaner visualization
non_zero_indices = [i for i, v in enumerate(Values) if v > 0]
Values = [Values[i] for i in non_zero_indices]
Category = [Category[i] for i in non_zero_indices]

# Sort both Values and Category by Values in descending order
sorted_pairs = sorted(zip(Values, Category), reverse=True)
Values, Category = zip(*sorted_pairs)

# Function to generate random background color
def random_background_color():
    return (random.random(), random.random(), random.random())

# Generate random RGB colors for each category
colors_random = [(random.random(), random.random(), random.random()) for _ in Category]

# Create figure with random background color
bg_color = random_background_color()
plt.figure(figsize=(10, 8), facecolor=bg_color)

# Create pie chart without labels inside
wedges, texts, autotexts = plt.pie(Values, 
                                    labels=None,  # No labels inside
                                    autopct='%1.1f%%', 
                                    colors=colors_random, 
                                    startangle=90, 
                                    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
                                    pctdistance=0.85)  # Adjust percentage distance from center

# Add title
plt.title('Energy mix in Kenya', fontsize=15, color='white')

# Ensure percentage text is visible
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')

# Create legend with percentages outside the pie
total = sum(Values)
legend_labels = [f'{cat}: {val/total*100:.1f}%' for cat, val in zip(Category, Values)]
plt.legend(wedges, legend_labels, 
          title="Categories",
          loc="center left", 
          bbox_to_anchor=(1, 0, 0.5, 1),
          fontsize=10,
          title_fontsize=12,
          frameon=True,
          fancybox=True,
          shadow=True)

plt.tight_layout()
#plt.savefig('energy_mix_pie_chart3.png', bbox_inches='tight')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

Values = [0.183333333, 3.406666667, 1.04, 5.396666667, 0.291666667, 1.806666667]
Category = ['Bioenergy', 'Hydro', 'Other_fossil', 'Geothermal and other renewables', 'Solar', 'Wind']

# Remove zero values
non_zero_indices = [i for i, v in enumerate(Values) if v > 0]
Values = [Values[i] for i in non_zero_indices]
Category = [Category[i] for i in non_zero_indices]

# Sort both Values and Category by Values in descending order
sorted_pairs = sorted(zip(Values, Category), reverse=True)
Values, Category = zip(*sorted_pairs)

def random_background_color():
    return (random.random(), random.random(), random.random())

colors_random = [(random.random(), random.random(), random.random()) for _ in Category]

bg_color = random_background_color()
plt.figure(figsize=(12, 10), facecolor=bg_color)

# Create pie chart
# labeldistance=0.5 to 0.7 keeps text inside
wedges, texts = plt.pie(Values, 
                        labels=Category,            
                        labeldistance=0.65,          
                        autopct=None,               
                        colors=colors_random, 
                        startangle=90, 
                        wedgeprops={'edgecolor': 'white', 'linewidth': 1.5},
                        textprops={'color': 'white', 'fontsize': 9, 'fontweight': 'bold'})

# --- ROTATION LOGIC ---
for text in texts:
    text.set_rotation(60)           # Tilt text 60 degrees
    text.set_horizontalalignment('center')
    text.set_verticalalignment('center')
    # Optional: Wrap text for very long category names
    text.set_wrap(True)

plt.title('Energy mix in Kenya(average in past years)', fontsize=18, color='black', fontweight='bold', pad=30)

# Legend remains for clarity since text is now tilted
total = sum(Values)
legend_labels = [f'{cat}: {val/total*100:.1f}%' for cat, val in zip(Category, Values)]
plt.legend(wedges, legend_labels, 
          title="Energy Sources",
          loc="center left", 
          bbox_to_anchor=(1, 0, 0.5, 1),
          fontsize=10,
          frameon=True,
          shadow=True)

plt.tight_layout()
plt.show()
```
